```python
qc = QuantumCircuit(4)

# Create initial entanglement between Q0 and Q1, and Q2 and Q3
qc.h(0)
qc.cx(0, 1)
qc.h(2)
qc.cx(2, 3)
qc.barrier()
# Bell state measurement on Q1 and Q2
qc.cx(1, 2)
qc.h(1)
#qc.measure([1, 2], [0, 1])
qc.barrier()
# Apply corrective operations based on the measurement outcomes
qc.cz(0, 1)
qc.cx(2, 3)
print(qc)
```

```python
from qiskit.visualization import plot_state_qsphere
plot_state_qsphere(qc)
```
